```python
import numpy as np
import matplotlib.pyplot as plt

import distributions as dist
import MetropolisHastings as mh
import langevin as lan
import hamiltonian as ham
```

```python
#Define distributions
#Test distributions
#####################################################

#Gaussian
gaussian = dist.Gaussian(np.array([0,0]), np.diag([1,1]))

X = np.linspace(-3,3,100)
Y = np.linspace(-3,3,100)
z1 = [[gaussian([x,y]) for x in X] for y in Y]

#Multimodal
double = dist.Multimodal(2,offset=.8)

X = np.linspace(-5,5,100)
Y = np.linspace(-5,5,100)
z2 = [[double([x,y]) for x in X] for y in Y]

#Banana
banana = dist.Banana()

X = np.linspace(-4,4,100)
Y = np.linspace(-4,4,100)
z3 = [[banana([x,y]) for x in X] for y in Y]

#Donut
donut = dist.Donut()

X = np.linspace(-10,10,100)
Y = np.linspace(-10,10,100)
z4 = [[donut([x,y]) for x in X] for y in Y]

#Multimodal
multimodal = dist.Multimodal(5,offset=3.5)

X = np.linspace(-7,7,100)
Y = np.linspace(-7,7,100)
z5 = [[multimodal([x,y]) for x in X] for y in Y]
```

```python
#Define Samplers for each distribution
plt.figure(figsize=(20,5))

#Plot on gaussian
metropolis = mh.MetropolisHastings(np.array([5,5]), gaussian, burn=200)
mh_burn, mh_samples, mh_accepted, mh_rejected = metropolis(3000, all=True)

plt.subplot(1,4,1)
plt.contour(X,Y,z1)
plt.title('Gaussian')

plt.plot(mh_burn[:,0], mh_burn[:,1])
plt.scatter(mh_samples[mh_accepted,0], mh_samples[mh_accepted,1], color='g')
plt.scatter(mh_samples[mh_rejected,0], mh_samples[mh_rejected,1], color='r')

#plot on double
bimodal = mh.MetropolisHastings(np.array([5,5]), double, burn=200)
mh_burn, mh_samples, mh_accepted, mh_rejected = bimodal(3000, all=True)

plt.subplot(1,4,2)
plt.contour(X,Y,z2)
plt.title('Bimodal Gaussian')

plt.plot(mh_burn[:,0], mh_burn[:,1])
plt.scatter(mh_samples[mh_accepted,0], mh_samples[mh_accepted,1], color='g')
plt.scatter(mh_samples[mh_rejected,0], mh_samples[mh_rejected,1], color='r')

#plot on banana
bananamh = mh.MetropolisHastings(np.array([15,15]), banana, burn=200)
mh_burn, mh_samples, mh_accepted, mh_rejected = bananamh(3000, all=True)

#plot on donut
donutmh = mh.MetropolisHastings(np.array([5,5]), donut, burn=200)
mh_burn, mh_samples, mh_accepted, mh_rejected = donutmh(3000, all=True)

plt.subplot(1,4,3)
plt.contour(X,Y,z4)
plt.title('Donut')

plt.plot(mh_burn[:,0], mh_burn[:,1])
plt.scatter(mh_samples[mh_accepted,0], mh_samples[mh_accepted,1], color='g')
plt.scatter(mh_samples[mh_rejected,0], mh_samples[mh_rejected,1], color='r')

#plot on multimodal
multimodalmh = mh.MetropolisHastings(np.array([5,5]), multimodal, burn=200)
mh_burn, mh_samples, mh_accepted, mh_rejected = multimodalmh(3000, all=True)

plt.subplot(1,4,4)
plt.contour(X,Y,z5)
plt.title('Multimodal')

plt.plot(mh_burn[:,0], mh_burn[:,1])
plt.scatter(mh_samples[mh_accepted,0], mh_samples[mh_accepted,1], color='g')
plt.scatter(mh_samples[mh_rejected,0], mh_samples[mh_rejected,1], color='r')
```

```python
#Langevin MCMC
#Define Samplers for each distribution
plt.figure(figsize=(20,5))

#Plot on gaussian
gaussianmh = lan.LangevinMonteCarlo(np.array([5,5]), gaussian, burn=200)
lan_burn, lan_samples, lan_accepted, lan_rejected = gaussianmh(2000, all=True)

plt.subplot(1,4,1)
plt.contour(X,Y,z1)
plt.title('Gaussian')

plt.plot(lan_burn[:,0], lan_burn[:,1])
plt.scatter(lan_samples[lan_accepted,0], lan_samples[lan_accepted,1], color='g')
plt.scatter(lan_samples[lan_rejected,0], lan_samples[lan_rejected,1], color='r')

#plot on double
bimodal = lan.LangevinMonteCarlo(np.array([5,5]), double, burn=200)
lan_burn, lan_samples, lan_accepted, lan_rejected = bimodal(2000, all=True)

plt.subplot(1,4,1)
plt.contour(X,Y,z2)
plt.title('Bimodal Gaussian')

plt.plot(lan_burn[:,0], lan_burn[:,1])
plt.scatter(lan_samples[lan_accepted,0], lan_samples[lan_accepted,1], color='g')
plt.scatter(lan_samples[lan_rejected,0], lan_samples[lan_rejected,1], color='r')

#plot on donut
donutlan = lan.LangevinMonteCarlo(np.array([5,5]), donut, burn=200)
lan_burn, lan_samples, lan_accepted, lan_rejected = donutlan(2000, all=True)

plt.subplot(1,4,1)
plt.contour(X,Y,z3)
plt.title('Donut')

plt.plot(lan_burn[:,0], lan_burn[:,1])
plt.scatter(lan_samples[lan_accepted,0], lan_samples[lan_accepted,1], color='g')
plt.scatter(lan_samples[lan_rejected,0], lan_samples[lan_rejected,1], color='r')

#plot on multimodal
donutlan = lan.LangevinMonteCarlo(np.array([5,5]), donut, burn=200)
lan_burn, lan_samples, lan_accepted, lan_rejected = donutlan(2000, all=True)

plt.subplot(1,4,1)
plt.contour(X,Y,z3)
plt.title('Donut')

plt.plot(lan_burn[:,0], lan_burn[:,1])
plt.scatter(lan_samples[lan_accepted,0], lan_samples[lan_accepted,1], color='g')
plt.scatter(lan_samples[lan_rejected,0], lan_samples[lan_rejected,1], color='r')
```
